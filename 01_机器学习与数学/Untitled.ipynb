```python
# S = 1/0! + 1/1! + 1/2! + 1/3! + 1/4! + ...... + 1/n!
s = 0
n = 8
for i in range(1, n+1):
    for j in range(1, i+1):
        s += 1/j
print(s)
```

```python
# Python Code 示例 1
import math
import matplotlib.pylab as plt

if __name__ == '__main__':
    x = [float(i)/100.0 for i in range(1, 300)]
    y = [math.log(i) for i in x]
    plt.plot(x, y, 'r-', linewidth=3, label='log Curve')  # 普通线形图，plt.plot(x，y，颜色线型，线宽，标签)
    a = [x[20], x[175]]
    b = [y[20], y[175]]
    plt.plot(a, b, 'g-', linewidth=2)
    plt.plot(a, b, 'b*', markersize=15, alpha=0.75)
    plt.legend(loc='upper left')  # 图例位置
    plt.grid(True)  # 网格
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.show()
```

```python
# Python Code 示例 2
import numpy as np
if __name__ == '__main__':
    u = np.random.uniform(0.0, 1.0, 10000)
    plt.hist(u, 80, facecolor='g', alpha=0.75)  # 柱状图
    plt.grid(True)
    plt.show()
    
    times = 10000
    for time in range(times):
        u += np.random.uniform(0.0, 1.0, 10000)
    print(len(u))
    u /= times
    print(len(u))
    plt.hist(u, 80, facecolor='g', alpha=0.75)
    plt.grid(True)
    plt.show()
```

```python
# y = x ** 2
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams['font.sans-serif'] = ['SimHei']  # 指定默认字体
plt.rcParams['axes.unicode_minus'] = False  # 解决保存图像是负号'-'显示为方块的问题

x = np.linspace(-1, 1, 100)
y1 = x ** 2  # 凸函数，图形和自然语言不符，是因为 y1=x**2 的导数为 y'=2x，二阶导数 y''=2，是正数，那么二阶导数表述图像的凹凸性，所以是凸函数。
plt.plot(x, y1, label='凸函数')
plt.legend()
plt.show()

y2 = -x ** 2  # 凹函数，是因为 y2=-x**2 的导数为 y'=-2x，二阶导数 y''=-2，是负数，所以叫凹函数。
plt.plot(x, y2, label='凹函数')  
plt.legend(loc='upper left')
plt.show()
```

```python
# f(x) = x ** x, x > 0，x 的最小值
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams['font.sans-serif'] = ['SimHei']  # 指定默认字体
plt.rcParams['axes.unicode_minus'] = False  # 解决保存图像是负号'-'显示为方块的问题

x = np.linspace(0, 3, 3000)
y = x ** x
plt.plot(x, y)
plt.show()
```
